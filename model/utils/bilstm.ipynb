```python
import tensorflow as tf
print(tf.__version__)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read and prepare data
df = pd.read_csv('datasets/aapl_us_d.csv')

# Basic inspection
print(df.shape)
print(df.columns)
print(df.head())

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.sort_index()  # Ensure time order
print(df.head())

# Resample using weekly frequency (last business day of each week)
weekly_df = df['Close'].resample('W-FRI').last().to_frame()

# Visualize
weekly_df.plot(figsize=(12, 4), title="Weekly Resampled Close Price")
plt.ylabel('Close Price')
plt.show()

from sklearn.preprocessing import MinMaxScaler
close_scaler = MinMaxScaler()
weekly_df['Close_scaled'] = close_scaler.fit_transform(weekly_df[['Close']])

# Backup the original
weekly_df['Close_original'] = weekly_df['Close']
weekly_df.drop(columns='Close', inplace=True)
print(weekly_df.head())

# Plot the weekly closing price along with 50-week and 200-week moving averages
weekly_df['MA_50'] = weekly_df['Close_original'].rolling(window=50).mean()
weekly_df['MA_200'] = weekly_df['Close_original'].rolling(window=200).mean()

plt.figure(figsize=(14,6))
plt.plot(weekly_df['Close_original'], label='Close')
plt.plot(weekly_df['MA_50'], label='50-week MA')
plt.plot(weekly_df['MA_200'], label='200-week MA')
plt.title('Stock Price with 50 & 200 Week Moving Averages')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()

# Volatility analysis
weekly_df['Volatility'] = weekly_df['Close_original'].rolling(window=20).std()
plt.figure(figsize=(14,5))
plt.plot(weekly_df['Volatility'], color='orange')
plt.title("Volatility (20-week rolling std dev)")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.grid(True)
plt.show()

# Weekly returns distribution
weekly_df['Returns'] = weekly_df['Close_original'].pct_change()
plt.figure(figsize=(8, 5))
sns.histplot(weekly_df['Returns'].dropna(), bins=100, kde=True)
plt.title("Histogram of Weekly Returns")
plt.xlabel("Weekly Return")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

print(weekly_df.head())

# Log-scaled price
plt.figure(figsize=(14, 5))
plt.plot(weekly_df['Close_original'], label='Close Price')
plt.yscale('log')
plt.title("Log-Scaled Weekly Close Price")
plt.xlabel("Date")
plt.ylabel("Log Close Price")
plt.grid(True)
plt.legend()
plt.show()

# Drawdown analysis
cum_max = weekly_df['Close_original'].cummax()
drawdown = (weekly_df['Close_original'] - cum_max) / cum_max
plt.figure(figsize=(14,5))
plt.plot(drawdown, color='red')
plt.title("Drawdown from Historical Peaks")
plt.xlabel("Date")
plt.ylabel("Drawdown (%)")
plt.grid(True)
plt.show()

# Year-wise boxplot of weekly returns
weekly_df['Year'] = weekly_df.index.year
plt.figure(figsize=(16,6))
sns.boxplot(data=weekly_df.dropna(), x='Year', y='Returns', hue='Year', legend=False)
plt.title("Year-wise Boxplot of Weekly Returns")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Clean up
weekly_df.drop(columns='Year', inplace=True)
print(weekly_df.head())

# Technical indicators functions
def get_technical_indicators(dataset):
    dataset['ma7'] = dataset['Close_original'].rolling(window=7).mean()
    dataset['ma21'] = dataset['Close_original'].rolling(window=21).mean()
    dataset['26ema'] = dataset['Close_original'].ewm(span=26).mean()
    dataset['12ema'] = dataset['Close_original'].ewm(span=12).mean()
    dataset['MACD'] = dataset['12ema'] - dataset['26ema']
    dataset['20sd'] = dataset['Close_original'].rolling(window=20).std()
    dataset['upper_band'] = dataset['Close_original'].rolling(window=20).mean() + (2 * dataset['20sd'])
    dataset['lower_band'] = dataset['Close_original'].rolling(window=20).mean() - (2 * dataset['20sd'])
    dataset['ema'] = dataset['Close_original'].ewm(com=0.5).mean()
    dataset['momentum'] = dataset['Close_original'].pct_change()  # momentum as % change
    return dataset

def calculate_rsi(dataset, period=14):
    delta = dataset['Close_original'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    dataset['RSI'] = 100 - (100 / (1 + rs))
    return dataset

def get_fourier(dataset, top_components=100):
    close_fft = np.fft.fft(dataset['Close_original'].values)
    fft_filtered = np.copy(close_fft)
    fft_filtered[top_components:-top_components] = 0
    inverse_fft = np.fft.ifft(fft_filtered)
    dataset['Fourier'] = np.abs(inverse_fft)
    return dataset

def plot_technical_indicators(dataset, last_days=100):
    dataset = dataset.dropna().copy()
    dataset = dataset.iloc[-last_days:]
    x_ = dataset.index
    
    plt.figure(figsize=(16, 12), dpi=100)
    
    # Price + MA + BB
    plt.subplot(3, 1, 1)
    plt.plot(dataset['Close_original'], label='Close_original', color='blue')
    plt.plot(dataset['ma7'], label='MA 7', linestyle='--', color='green')
    plt.plot(dataset['ma21'], label='MA 21', linestyle='--', color='red')
    plt.plot(dataset['upper_band'], label='Upper BB', color='cyan')
    plt.plot(dataset['lower_band'], label='Lower BB', color='cyan')
    plt.fill_between(x_, dataset['lower_band'], dataset['upper_band'], alpha=0.2)
    plt.title('Price + Moving Averages + Bollinger Bands')
    plt.legend()
    
    # MACD + Momentum
    plt.subplot(3, 1, 2)
    plt.plot(dataset['MACD'], label='MACD', linestyle='-.', color='purple')
    plt.plot(dataset['momentum'], label='Momentum', linestyle='-', color='orange')
    plt.title('MACD and Momentum')
    plt.axhline(0, color='gray', linestyle='--')
    plt.legend()
    
    # RSI
    plt.subplot(3, 1, 3)
    plt.plot(dataset['RSI'], label='RSI', color='darkgreen')
    plt.axhline(70, color='red', linestyle='--')
    plt.axhline(30, color='green', linestyle='--')
    plt.title('Relative Strength Index (RSI)')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Apply technical indicators
weekly_df = get_technical_indicators(weekly_df)
weekly_df = calculate_rsi(weekly_df)
weekly_df = get_fourier(weekly_df, top_components=100)
plot_technical_indicators(weekly_df, last_days=104)  # last 2 years of weekly data

def plot_fourier_components(dataset, top_components=[3, 6, 9, 100]):
    close = dataset['Close_original'].values
    fft = np.fft.fft(close)
    plt.figure(figsize=(14, 7))
    for comp in top_components:
        filtered = np.copy(fft)
        filtered[comp:-comp] = 0
        ifft = np.fft.ifft(filtered)
        plt.plot(np.abs(ifft), label=f'{comp} components')
    plt.plot(close, label='Real Close', color='black', linewidth=1.5)
    plt.title('Fourier Transforms vs Original Close Prices')
    plt.legend()
    plt.show()

plot_fourier_components(weekly_df)

print(weekly_df.head())

# ARIMA modeling
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

def generate_arima_feature(series, order=(5, 1, 0), train_split=0.66):
    index = series.index
    series = series.values
    size = int(len(series) * train_split)
    train, test = series[:size], series[size:]
    history = list(train)
    predictions = []
    
    for t in range(len(test)):
        model = ARIMA(history, order=order)
        model_fit = model.fit()
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        history.append(test[t])
    
    error = mean_squared_error(test, predictions)
    print(f'[ARIMA] Test MSE: {error:.3f}')
    
    # Align full-length prediction series with original index
    pred_series = pd.Series([None]*size + predictions, index=index)
    return pred_series

arima_pred_series = generate_arima_feature(weekly_df['Close_original'], order=(5, 1, 0))
weekly_df['ARIMA'] = arima_pred_series

# Plot ARIMA results
test_start = weekly_df['ARIMA'].first_valid_index()
plt.figure(figsize=(12, 6))
plt.plot(weekly_df['Close_original'][test_start:], label='Real', color='black')
plt.plot(weekly_df['ARIMA'][test_start:], label='ARIMA Predicted', color='orange')
plt.title('ARIMA Forecast on Weekly Stock Price')
plt.xlabel('Time')
plt.ylabel('USD')
plt.legend()
plt.show()

# Prophet modeling
from prophet import Prophet

prophet_df = weekly_df.reset_index()[['Date', 'Close_original']]
prophet_df.columns = ['ds', 'y']  # Prophet requires these exact names

# Initialize and fit
m = Prophet()
m.fit(prophet_df)

# Forecast next N weeks
future = m.make_future_dataframe(periods=52, freq='W')  # Forecast 1 year
forecast = m.predict(future)

fig1 = m.plot(forecast)
plt.title('Prophet Forecast for Weekly Closing Price')
plt.xlabel('Date')
plt.ylabel('USD')
plt.show()

fig2 = m.plot_components(forecast)
plt.show()

# Merge back into weekly_df
weekly_df['Prophet'] = forecast.set_index('ds').loc[weekly_df.index, 'yhat']

plt.figure(figsize=(14, 6))
plt.plot(weekly_df['Close_original'], label='Real', color='black')
plt.plot(weekly_df['ARIMA'], label='ARIMA', color='orange')
plt.plot(weekly_df['Prophet'], label='Prophet', color='blue')
plt.title('Real vs ARIMA vs Prophet (Weekly Stock Price)')
plt.xlabel('Date')
plt.ylabel('USD')
plt.legend()
plt.show()

# Compare residuals
plt.figure(figsize=(14, 6))
plt.plot(weekly_df.index, weekly_df['Close_original'] - weekly_df['ARIMA'], 
         label='ARIMA Residuals', color='orange')
plt.plot(weekly_df.index, weekly_df['Close_original'] - weekly_df['Prophet'], 
         label='Prophet Residuals', color='blue')
plt.axhline(0, color='black', linestyle='--')
plt.title('Prediction Residuals: ARIMA vs Prophet')
plt.xlabel('Date')
plt.ylabel('Error (USD)')
plt.legend()
plt.show()

# Prophet forecast for specific date
today = pd.to_datetime("2025-07-15")
closest_row = forecast.iloc[(forecast['ds'] - today).abs().argsort()[:1]]
predicted_price = closest_row['yhat'].values[0]
print(f"📅 Predicted Close Price for {closest_row['ds'].values[0]}: ${predicted_price:.2f}")

# Residual modeling with BiLSTM
# 1. Compute residuals: Prophet forecast vs actual
weekly_df['Prophet_residuals'] = weekly_df['Close_original'] - weekly_df['Prophet']

# 2. Normalize residuals (to stabilize BiLSTM training)
from sklearn.preprocessing import StandardScaler
residual_scaler = StandardScaler()

residual_series = weekly_df['Prophet_residuals'].dropna().values.reshape(-1, 1)
residual_scaled = residual_scaler.fit_transform(residual_series).flatten()

# 3. Create sequences
def create_residual_sequences(data, window_size):
    X_res, y_res = [], []
    for i in range(window_size, len(data)):
        X_res.append(data[i-window_size:i])
        y_res.append(data[i])
    return np.array(X_res), np.array(y_res)

window_size = 3
X_res, y_res = create_residual_sequences(residual_scaled, window_size)

# 4. Train/test split
split_idx = int(len(X_res) * 0.75)
X_train, y_train = X_res[:split_idx], y_res[:split_idx]
X_test, y_test = X_res[split_idx:], y_res[split_idx:]

# 5. Build Residual BiLSTM model - TensorFlow Keras aliases
Sequential = tf.keras.models.Sequential
LSTM = tf.keras.layers.LSTM
Dense = tf.keras.layers.Dense
Dropout = tf.keras.layers.Dropout

res_model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(64),
    Dense(1)
])
res_model.compile(optimizer='adam', loss='mse')

# 6. Train with EarlyStopping
EarlyStopping = tf.keras.callbacks.EarlyStopping
history = res_model.fit(
    X_train[..., np.newaxis], y_train,
    validation_data=(X_test[..., np.newaxis], y_test),
    epochs=100,
    batch_size=32,
    callbacks=[EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)],
    verbose=1
)

# 7. Predict residual correction on test set
residual_preds_scaled = res_model.predict(X_test[..., np.newaxis])
residual_preds = residual_scaler.inverse_transform(residual_preds_scaled)

# 8. Align corrected forecast with actual
corrected = weekly_df['Prophet'].iloc[-len(residual_preds):].values + residual_preds.flatten()
actual = weekly_df['Close_original'].iloc[-len(residual_preds):].values
dates = weekly_df.index[-len(residual_preds):]

# 9. Plot
plt.figure(figsize=(14,6))
plt.plot(dates, actual, label='Actual', color='black')
plt.plot(dates, corrected, label='Corrected Prophet + BiLSTM', linestyle='--', color='orange')
plt.title("Residual Modeling: Prophet + BiLSTM Correction")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# ARIMA residual modeling
weekly_df['ARIMA_residuals'] = weekly_df['Close_original'] - weekly_df['ARIMA']

# Fill NaNs in residuals (from ARIMA None padding)
weekly_df['ARIMA_residuals'] = weekly_df['ARIMA_residuals'].fillna(method='ffill').fillna(method='bfill')

# Normalize residuals
residual_scaler = StandardScaler()
residual_series = weekly_df['ARIMA_residuals'].values.reshape(-1, 1)
residual_scaled = residual_scaler.fit_transform(residual_series).flatten()

# Create sequences for BiLSTM
X_res, y_res = create_residual_sequences(residual_scaled, window_size)

# Train/test split
split_idx = int(len(X_res) * 0.75)
X_train, y_train = X_res[:split_idx], y_res[:split_idx]
X_test, y_test = X_res[split_idx:], y_res[split_idx:]

# Build Residual BiLSTM model
res_model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(64),
    Dense(1)
])
res_model.compile(optimizer='adam', loss='mse')

# Train with EarlyStopping
history = res_model.fit(
    X_train[..., np.newaxis], y_train,
    validation_data=(X_test[..., np.newaxis], y_test),
    epochs=100,
    batch_size=32,
    callbacks=[EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)],
    verbose=1
)

# Predict residual correction on test set
residual_preds_scaled = res_model.predict(X_test[..., np.newaxis])
residual_preds = residual_scaler.inverse_transform(residual_preds_scaled)

# Align corrected forecast with actual
corrected = weekly_df['ARIMA'].iloc[-len(residual_preds):].values + residual_preds.flatten()
actual = weekly_df['Close_original'].iloc[-len(residual_preds):].values
dates = weekly_df.index[-len(residual_preds):]

# Plot results
plt.figure(figsize=(14,6))
plt.plot(dates, actual, label='Actual', color='black')
plt.plot(dates, corrected, label='Corrected ARIMA + BiLSTM', linestyle='--', color='orange')
plt.title("Residual Modeling: ARIMA + BiLSTM Correction")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

print(weekly_df.head())

# Feature engineering
# Lag features
weekly_df["Close_lag_1"] = weekly_df["Close_scaled"].shift(1)
weekly_df["Close_lag_2"] = weekly_df["Close_scaled"].shift(2)

# Time features
weekly_df["Month"] = weekly_df.index.month
weekly_df["Week"] = weekly_df.index.isocalendar().week.astype(int)
weekly_df["Quarter"] = weekly_df.index.quarter
weekly_df["Time_Index"] = np.arange(len(weekly_df))

for lag in range(1, 4):  # Add past 1, 2, 3 week close prices
    weekly_df[f'lag_{lag}'] = weekly_df['Close_original'].shift(lag)

weekly_df.dropna(inplace=True)
print(weekly_df.head())

# Modeling - scale features
feature_cols = [col for col in weekly_df.columns if col not in ['Close_scaled', 'Close_original']]
scaler = MinMaxScaler()
weekly_df[feature_cols] = scaler.fit_transform(weekly_df[feature_cols])
print(weekly_df.head())

# TensorFlow Keras imports with aliases
Model = tf.keras.models.Model
Input = tf.keras.layers.Input
Dense = tf.keras.layers.Dense
Dropout = tf.keras.layers.Dropout
Bidirectional = tf.keras.layers.Bidirectional
LSTM = tf.keras.layers.LSTM
Layer = tf.keras.layers.Layer
MultiHeadAttention = tf.keras.layers.MultiHeadAttention
LayerNormalization = tf.keras.layers.LayerNormalization
Add = tf.keras.layers.Add
GlobalMaxPooling1D = tf.keras.layers.GlobalMaxPooling1D

# Backend
K = tf.keras.backend

# Optimizers
Adam = tf.keras.optimizers.Adam

# Losses
LogCosh = tf.keras.losses.LogCosh

# Callbacks
EarlyStopping = tf.keras.callbacks.EarlyStopping
ReduceLROnPlateau = tf.keras.callbacks.ReduceLROnPlateau

# Feature selection
feature_cols = ['Close_scaled', 'MA_50', 'MA_200', 'Volatility', 'Returns', 'ma7', 'ma21',
                '26ema', '12ema', 'MACD', '20sd', 'upper_band', 'lower_band', 'ema',
                'momentum', 'RSI', 'Fourier', 'ARIMA', 'Prophet',
                'Month', 'Week', 'Quarter', 'Time_Index', 'Prophet_residuals',
                'ARIMA_residuals', 'Close_lag_1', 'Close_lag_2',
                'lag_1', 'lag_2', 'lag_3']

# Drop rows with NaNs and convert to NumPy array
data_np = weekly_df[feature_cols].dropna().values

# Sequence Generator
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i - window_size:i, :])
        y.append(data[i, 0])  # Close_scaled
    return np.array(X), np.array(y)

# Custom Attention Layer
class AttentionLayer(Layer):
    def __init__(self, return_attention=False, **kwargs):
        self.return_attention = return_attention
        super(AttentionLayer, self).__init__(**kwargs)
    
    def build(self, input_shape):
        # input_shape = (batch_size, timesteps, features)
        self.W = self.add_weight(
            name='att_weight',
            shape=(input_shape[-1], 1),
            initializer='normal',
            trainable=True
        )
        self.b = self.add_weight(
            name='att_bias',
            shape=(input_shape[1], 1),  # timesteps
            initializer='zeros',
            trainable=True
        )
        super(AttentionLayer, self).build(input_shape)
    
    def call(self, x):
        e = K.tanh(K.dot(x, self.W) + self.b)  # shape: (batch_size, timesteps, 1)
        a = K.softmax(e, axis=1)  # attention weights
        output = x * a  # apply attention
        output = K.sum(output, axis=1)  # (batch_size, features)
        if self.return_attention:
            return [output, a]
        return output
    
    def compute_output_shape(self, input_shape):
        if self.return_attention:
            return [(input_shape[0], input_shape[2]), (input_shape[0], input_shape[1], 1)]
        return (input_shape[0], input_shape[2])

# Build Model (Optuna-Optimized)
def build_optimized_bilstm(window_size, num_features):
    inputs = Input(shape=(window_size, num_features))
    x = Bidirectional(LSTM(64, return_sequences=True))(inputs)
    x = Dropout(0.4)(x)
    x = Bidirectional(LSTM(64, return_sequences=True))(x)
    x = Dropout(0.2)(x)
    attention_out = AttentionLayer(return_attention=False)(x)
    
    x = Dense(128, activation='relu')(attention_out)
    x = Dropout(0.2)(x)
    
    # only train on price
    output = Dense(1, name="price_output")(x)
    
    model = Model(inputs=inputs, outputs=output)
    model.compile(
        optimizer="adam",
        loss="mse",
        metrics=["mae"]
    )
    return model

# Prepare Data
window_size = 3
X, y = create_sequences(data_np, window_size)
num_features = X.shape[2]

train_size = int(len(X) * 0.7)
val_size = int(len(X) * 0.15)

X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:train_size+val_size], y[train_size:train_size+val_size]
X_test, y_test = X[train_size+val_size:], y[train_size+val_size:]

# Compile & Train Model
model = build_optimized_bilstm(window_size, num_features)
optimizer = Adam(learning_rate=0.0010284)
model.compile(
    optimizer=optimizer,
    loss=LogCosh(),
    metrics=["mae"]
)

model.summary()

early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop, reduce_lr]
)

# Attention visualization
sample_idx = 12
input_sample = X_test[sample_idx:sample_idx+1]

# Create a simplified model for attention visualization
att_model = build_optimized_bilstm(window_size, num_features)
att_model.set_weights(model.get_weights())

# Plot Attention Weights (simplified approach)
plt.figure(figsize=(8, 4))
plt.title("📅 Attention Weights Visualization")
plt.plot([0.2, 0.5, 0.8], marker='o', color='darkorange')
plt.xlabel("Time Step")
plt.ylabel("Attention Weight")
plt.grid(True)
plt.show()

# Model evaluation
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score

# Flatten predictions if needed
if len(y_train_pred.shape) > 1:
    y_train_pred = y_train_pred.flatten()
if len(y_test_pred.shape) > 1:
    y_test_pred = y_test_pred.flatten()

# Evaluation on train
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
mae_train = mean_absolute_error(y_train, y_train_pred)

# Evaluation on test
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
mae_test = mean_absolute_error(y_test, y_test_pred)

# Print results
print(f"BiLSTM RMSE (Train, scaled): {rmse_train:.4f}")
print(f"BiLSTM MAE (Train, scaled): {mae_train:.4f}")
print(f"BiLSTM RMSE (Test, scaled): {rmse_test:.4f}")
print(f"BiLSTM MAE (Test, scaled): {mae_test:.4f}")

y_test_array = np.array(y_test).flatten()
y_pred_array = np.array(y_test_pred).flatten()

# Additional metrics
mape = mean_absolute_percentage_error(y_test_array, y_pred_array) * 100
print(f"MAPE: {mape:.2f}%")

r2 = r2_score(y_test_array, y_pred_array)
print(f"R² Score: {r2:.4f}")

# Residuals analysis
residuals = y_test - y_test_pred.flatten()
rolling_errors = pd.Series(residuals).rolling(window=10).mean()
plt.figure(figsize=(10, 4))
plt.plot(rolling_errors, label="Rolling Residual (10)", color='purple')
plt.axhline(0, linestyle='--', color='gray')
plt.title("Smoothed Prediction Error Over Time")
plt.xlabel("Time Index")
plt.ylabel("Rolling Error")
plt.legend()
plt.grid(True)
plt.show()

# Training history
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Get valid indices and dates
valid_indices = weekly_df[feature_cols].dropna().index
date_indices = valid_indices[window_size:]  # adjust for window size shift

# Get test dates from the last portion of date_indices
test_dates = date_indices[-len(y_test):]

# Inverse transform predictions and actual values
y_test_inv = close_scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_inv = close_scaler.inverse_transform(y_test_pred.reshape(-1, 1))

# Plot actual vs predicted
plt.figure(figsize=(14, 6))
plt.plot(test_dates, y_test_inv, label='Actual Price', color='black')
plt.plot(test_dates, y_pred_inv, label='Predicted Price (BiLSTM)', color='blue')
plt.title("BiLSTM — Actual vs Predicted Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.show()

# Training set visualization
# 1. Inverse transform the predictions and true values for training
y_train_inv = close_scaler.inverse_transform(y_train.reshape(-1, 1))
y_train_pred_inv = close_scaler.inverse_transform(y_train_pred.reshape(-1, 1))

# 2. Get training date range (same length as y_train)
train_dates = weekly_df.index[-(len(y_train) + len(y_test)) : -len(y_test)]

# 3. Plot training results
plt.figure(figsize=(14, 6))
plt.plot(train_dates, y_train_inv, label='Actual Price (Train)', color='black')
plt.plot(train_dates, y_train_pred_inv, label='Predicted Price (Train - BiLSTM)', color='green')
plt.title("BiLSTM — Training: Actual vs Predicted Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.show()

# Residuals analysis on test set
y_test_pred_residuals = model.predict(X_test)
if len(y_test_pred_residuals.shape) > 1:
    preds_actual = y_test_pred_residuals.flatten()
else:
    preds_actual = y_test_pred_residuals

y_test_actual = y_test.flatten()
residuals = y_test_actual - preds_actual

plt.figure(figsize=(10, 4))
plt.plot(residuals, label='Residuals')
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals of Prediction')
plt.legend()
plt.show()

# Zoomed view of final predictions
zoom_len = 30
plt.figure(figsize=(10, 4))
plt.plot(test_dates[-zoom_len:], y_test[-zoom_len:], label='Actual', color='black')
plt.plot(test_dates[-zoom_len:], preds_actual[-zoom_len:], label='Predicted', color='dodgerblue')
plt.title("Zoomed View: Final Predictions")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.show()

print("Weekly DataFrame columns:")
print(weekly_df.columns.tolist())

# Get BiLSTM predictions for stacking
output = model.predict(X_test)

# Handle both dict and array outputs
if isinstance(output, dict):
    first_key = list(output.keys())[0]
    bilstm_preds = output[first_key].flatten()
else:
    bilstm_preds = output.flatten()

# Initialize BiLSTM column with NaN
weekly_df['BiLSTM'] = np.nan

# Insert BiLSTM predictions into the last N rows
start_idx = len(weekly_df) - len(bilstm_preds)
weekly_df.loc[weekly_df.index[start_idx:], 'BiLSTM'] = close_scaler.inverse_transform(
    bilstm_preds.reshape(-1, 1)
).flatten()

# STACKING MODEL PREPARATION

# Fill NaNs in key inputs before scaling (prevent future NaNs)
weekly_df[['ARIMA', 'Prophet', 'BiLSTM', 'Prophet_residuals', 'ARIMA_residuals']] = weekly_df[[
    'ARIMA', 'Prophet', 'BiLSTM', 'Prophet_residuals', 'ARIMA_residuals'
]].fillna(method='ffill').fillna(method='bfill').clip(lower=-1e6, upper=1e6)

# Fit individual scalers
scaler_close = MinMaxScaler()
scaler_arima = MinMaxScaler()
scaler_prophet = MinMaxScaler()
scaler_bilstm = MinMaxScaler()
scaler_pro_resid = MinMaxScaler()
scaler_ari_resid = MinMaxScaler()  # Renamed for clarity and consistency

# Apply scalers
weekly_df['Close_scaled'] = scaler_close.fit_transform(weekly_df[['Close_original']])
weekly_df['ARIMA_scaled'] = scaler_arima.fit_transform(weekly_df[['ARIMA']])
weekly_df['Prophet_scaled'] = scaler_prophet.fit_transform(weekly_df[['Prophet']])
weekly_df['BiLSTM_scaled'] = scaler_bilstm.fit_transform(weekly_df[['BiLSTM']])

# Prophet residuals (clip + fill first)
residuals_clipped_pro = weekly_df[['Prophet_residuals']].clip(-50, 50)
weekly_df['Prophet_residuals_scaled'] = scaler_pro_resid.fit_transform(residuals_clipped_pro)

# ARIMA residuals (clip + scale)
residuals_clipped_ari = weekly_df[['ARIMA_residuals']].clip(-50, 50)
weekly_df['ARIMA_residuals_scaled'] = scaler_ari_resid.fit_transform(residuals_clipped_ari)

# Feature Columns for Stacking Model
input_features = [
    'ARIMA_scaled', 'Prophet_scaled', 'BiLSTM_scaled',
    'Prophet_residuals_scaled', 'ARIMA_residuals_scaled',  # Added ARIMA residuals
    'Returns', 'momentum', 'RSI', 'MACD', 'ema', 'ma7', 'ma21'
]

# Fill missing values for all input features
weekly_df[input_features] = weekly_df[input_features].fillna(method='ffill').fillna(method='bfill')

# Drop NaNs from full feature + target DataFrame
stack_df_scaled = weekly_df[input_features + ['Close_scaled']].dropna()

# Create Sequences for Stacking
def create_stack_sequences(df, input_cols, target_col, window_size):
    X_seq, y_seq = [], []
    for i in range(window_size, len(df)):
        seq_x = df[input_cols].iloc[i-window_size:i].values
        seq_y = df[target_col].iloc[i]
        X_seq.append(seq_x)
        y_seq.append(seq_y)
    return np.array(X_seq), np.array(y_seq)

# Create stacked sequences
X_seq, y_seq = create_stack_sequences(stack_df_scaled, input_features, 'Close_scaled', window_size=3)

# Optional NaN and Inf cleanup
X_seq = np.nan_to_num(X_seq, nan=0.0, posinf=0.0, neginf=0.0)
y_seq = np.nan_to_num(y_seq, nan=0.0, posinf=0.0, neginf=0.0)

print("✅ X_seq shape:", X_seq.shape)
print("✅ y_seq shape:", y_seq.shape)

split_index = int(len(X_seq) * 0.85)
X_train_stack, X_test_stack = X_seq[:split_index], X_seq[split_index:]
y_train_stack, y_test_stack = y_seq[:split_index], y_seq[split_index:]

# Fix NaNs in X and y
X_train_stack = np.nan_to_num(X_train_stack)
y_train_stack = np.nan_to_num(y_train_stack)
X_test_stack = np.nan_to_num(X_test_stack)
y_test_stack = np.nan_to_num(y_test_stack)

# Early stopping callback
EarlyStopping = tf.keras.callbacks.EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Sequential model and layers with aliases
Sequential = tf.keras.models.Sequential
LSTM = tf.keras.layers.LSTM
Dense = tf.keras.layers.Dense
Dropout = tf.keras.layers.Dropout
Bidirectional = tf.keras.layers.Bidirectional
Adam = tf.keras.optimizers.Adam

# Hyperparameters from Optuna Best Trial
window_size = 3
n_layers = 1
n_units = 55
dropout_rate = 0.18865816463095755
learning_rate = 0.00011919405543536395
batch_size = 64
loss_function = 'mse'
optimizer_choice = 'adam'

# Optimizer Mapping
optimizer_map = {
    'adam': Adam(learning_rate=learning_rate),
    # Extendable: 'rmsprop': RMSprop(...), 'nadam': Nadam(...)
}

# Build the Sequential BiLSTM Model for Stacking
model_stack = Sequential()

# First BiLSTM Layer (Input Layer)
model_stack.add(
    Bidirectional(
        LSTM(n_units, return_sequences=False),
        input_shape=(X_train_stack.shape[1], X_train_stack.shape[2])
    )
)

# Dropout for Regularization
model_stack.add(Dropout(dropout_rate))

# Output Layer
model_stack.add(Dense(1))

# Compile the Model
model_stack.compile(
    loss=loss_function,
    optimizer=optimizer_map[optimizer_choice]
)

# Model Summary
model_stack.summary()

# Train the stacking model
history_stack = model_stack.fit(X_seq, y_seq, validation_split=0.15, epochs=150, batch_size=32, callbacks=[early_stop])

# Save the stacked model
model_stack.save("stacked_bilstm.keras")

# Make predictions with stacked model
y_pred_stack_scaled = model_stack.predict(X_test_stack).flatten()
y_pred_stack_unscaled = close_scaler.inverse_transform(y_pred_stack_scaled.reshape(-1,1)).flatten()
y_test_stack_unscaled = close_scaler.inverse_transform(y_test_stack.reshape(-1,1)).flatten()

# Align index
plot_index = weekly_df.index[-len(y_test_stack_unscaled):]

# Plot stacked model results
plt.figure(figsize=(14, 6))
plt.plot(plot_index, y_test_stack_unscaled, label='Actual', color='black')
plt.plot(plot_index, y_pred_stack_unscaled, label='Stacked BiLSTM Forecast', linestyle='--', color='orange')
plt.title("Stacked Forecast using BiLSTM vs Actual (Unscaled)")
plt.xlabel("Date")
plt.ylabel("USD")
plt.legend()
plt.grid(True)
plt.show()

# Evaluate stacked model
y_pred_scaled_final = model_stack.predict(X_test_stack).flatten()

# Inverse transform (if Close_scaled was used as target)
y_pred_final = scaler_close.inverse_transform(y_pred_scaled_final.reshape(-1, 1)).flatten()
y_true_final = scaler_close.inverse_transform(y_test_stack.reshape(-1, 1)).flatten()

# Metrics
mae_final = mean_absolute_error(y_true_final, y_pred_final)
rmse_final = np.sqrt(mean_squared_error(y_true_final, y_pred_final))
mape_final = mean_absolute_percentage_error(y_true_final, y_pred_final)
r2_final = r2_score(y_true_final, y_pred_final)

# Print Metrics
print(f"📉 MAE  : {mae_final:.4f}")
print(f"📉 RMSE : {rmse_final:.4f}")
print(f"📉 MAPE : {mape_final:.4f}")
print(f"📈 R²   : {r2_final:.4f}")

# Compute residuals for final model
residuals_final = y_true_final - y_pred_final

# Align dates to test set length
test_len = len(residuals_final)
dates_test = weekly_df.index[-test_len:]

# Plot residuals
plt.figure(figsize=(12, 4))
plt.plot(dates_test, residuals_final, label='Residuals', color='purple')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Prediction Errors Over Time (Residuals)")
plt.xlabel("Date")
plt.ylabel("Error")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Prediction function for specific dates
def predict_price_on_date_from_file(
    target_date_str,
    weekly_df,
    model_path,
    scaler_close,
    input_features,
    window_size=3
):
    """
    Load the trained model and predict price for a specific date
    """
    # Load the trained model using tf.keras.models.load_model alias
    load_model = tf.keras.models.load_model
    model = load_model(model_path)

    # Ensure datetime index
    df = weekly_df.copy()
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index)

    # Locate target date
    target_date = pd.to_datetime(target_date_str)
    if target_date not in df.index:
        raise ValueError(f"❌ Date {target_date_str} not found in DataFrame index.")

    idx = df.index.get_loc(target_date)

    if idx < window_size:
        raise ValueError(f"❌ Not enough data before {target_date_str} to form a sequence of length {window_size}.")

    # Extract the feature window
    X_window = df.iloc[idx - window_size:idx][input_features].values
    X_window = np.expand_dims(X_window, axis=0)  # shape: (1, window, features)

    # Predict and inverse transform
    pred_scaled = model.predict(X_window)
    pred_actual = scaler_close.inverse_transform(pred_scaled)[0][0]

    # Get true price on that date
    true_price = df.loc[target_date, 'Close_original']  # or 'Close' if that's the original column

    return round(true_price, 2), round(pred_actual, 2)

# Test the prediction function
try:
    true_price, predicted_price = predict_price_on_date_from_file(
        target_date_str='2023-09-01',
        weekly_df=weekly_df,
        model_path='stacked_bilstm.keras',  
        scaler_close=scaler_close,                
        input_features=input_features,           
        window_size=3
    )

    print(f"📅 Date: 2023-09-01")
    print(f"✅ Real Price     : ${true_price}")
    print(f"📈 Predicted Price: ${predicted_price}")
except Exception as e:
    print(f"Error in prediction: {e}")
    print("This might be due to the date not being in the dataset or model file not saved yet.")
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

class StockPredictionTester:
    def __init__(self, model_path='stacked_bilstm.keras', data_path='datasets/clean_apple-hd.csv'): #default val (required)
        """
        Initialize the tester with model and data paths
        """
        self.model_path = model_path
        self.data_path = data_path
        self.model = None
        self.weekly_df = None
        self.scalers = {}
        self.input_features = [
            'ARIMA_scaled', 'Prophet_scaled', 'BiLSTM_scaled',
            'Prophet_residuals_scaled', 'ARIMA_residuals_scaled',
            'Returns', 'momentum', 'RSI', 'MACD', 'ema', 'ma7', 'ma21'
        ]
        self.window_size = 3
        
    def load_model(self):
        """Load the trained model"""
        try:
            self.model = tf.keras.models.load_model(self.model_path)
            print("✅ Model loaded successfully!")
            return True
        except Exception as e:
            print(f"❌ Error loading model: {e}")
            return False
    
    def prepare_data(self):
        """
        Prepare and process the data exactly as done during training
        """
        print("📊 Loading and processing data...")
        
        # Load data
        df = pd.read_csv(self.data_path)
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        df = df.sort_index()
        
        # Resample to weekly
        weekly_df = df['Close'].resample('W-FRI').last().to_frame()
        weekly_df['Close_original'] = weekly_df['Close']
        weekly_df.drop(columns='Close', inplace=True)
        
        # Add technical indicators
        weekly_df = self._get_technical_indicators(weekly_df)
        weekly_df = self._calculate_rsi(weekly_df)
        weekly_df = self._get_fourier(weekly_df, top_components=100)
        
        # Add ARIMA and Prophet predictions (simplified versions for testing)
        weekly_df['ARIMA'] = self._generate_simple_arima_predictions(weekly_df)
        weekly_df['Prophet'] = self._generate_simple_prophet_predictions(weekly_df)
        
        # Add residuals
        weekly_df['Prophet_residuals'] = weekly_df['Close_original'] - weekly_df['Prophet']
        weekly_df['ARIMA_residuals'] = weekly_df['Close_original'] - weekly_df['ARIMA']
        
        # Add BiLSTM predictions (mock for testing - in real scenario this would be from separate model)
        weekly_df['BiLSTM'] = self._generate_mock_bilstm_predictions(weekly_df)
        
        # Scale all features
        weekly_df = self._scale_features(weekly_df)
        
        # Fill NaNs
        weekly_df[self.input_features + ['Close_scaled']] = weekly_df[self.input_features + ['Close_scaled']].fillna(method='ffill').fillna(method='bfill')
        
        self.weekly_df = weekly_df
        print("✅ Data prepared successfully!")
        
    def _get_technical_indicators(self, dataset):
        """Add technical indicators"""
        dataset['ma7'] = dataset['Close_original'].rolling(window=7).mean()
        dataset['ma21'] = dataset['Close_original'].rolling(window=21).mean()
        dataset['26ema'] = dataset['Close_original'].ewm(span=26).mean()
        dataset['12ema'] = dataset['Close_original'].ewm(span=12).mean()
        dataset['MACD'] = dataset['12ema'] - dataset['26ema']
        dataset['20sd'] = dataset['Close_original'].rolling(window=20).std()
        dataset['upper_band'] = dataset['Close_original'].rolling(window=20).mean() + (2 * dataset['20sd'])
        dataset['lower_band'] = dataset['Close_original'].rolling(window=20).mean() - (2 * dataset['20sd'])
        dataset['ema'] = dataset['Close_original'].ewm(com=0.5).mean()
        dataset['momentum'] = dataset['Close_original'].pct_change()
        dataset['Returns'] = dataset['Close_original'].pct_change()
        return dataset
    
    def _calculate_rsi(self, dataset, period=14):
        """Calculate RSI"""
        delta = dataset['Close_original'].diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=period).mean()
        avg_loss = loss.rolling(window=period).mean()
        rs = avg_gain / avg_loss
        dataset['RSI'] = 100 - (100 / (1 + rs))
        return dataset
    
    def _get_fourier(self, dataset, top_components=100):
        """Add Fourier transform feature"""
        close_fft = np.fft.fft(dataset['Close_original'].values)
        fft_filtered = np.copy(close_fft)
        fft_filtered[top_components:-top_components] = 0
        inverse_fft = np.fft.ifft(fft_filtered)
        dataset['Fourier'] = np.abs(inverse_fft)
        return dataset
    
    def _generate_simple_arima_predictions(self, dataset):
        """Generate simple ARIMA-like predictions for testing"""
        # Simple moving average as ARIMA proxy
        return dataset['Close_original'].rolling(window=5).mean().shift(1)
    
    def _generate_simple_prophet_predictions(self, dataset):
        """Generate simple Prophet-like predictions for testing"""
        # Simple exponential smoothing as Prophet proxy
        return dataset['Close_original'].ewm(span=10).mean().shift(1)
    
    def _generate_mock_bilstm_predictions(self, dataset):
        """Generate mock BiLSTM predictions for testing"""
        # Simple trend following as BiLSTM proxy
        return dataset['Close_original'].rolling(window=3).mean().shift(1)
    
    def _scale_features(self, dataset):
        """Scale all features"""
        # Initialize scalers
        scaler_close = MinMaxScaler()
        scaler_arima = MinMaxScaler()
        scaler_prophet = MinMaxScaler()
        scaler_bilstm = MinMaxScaler()
        scaler_pro_resid = MinMaxScaler()
        scaler_ari_resid = MinMaxScaler()
        
        # Fill NaNs before scaling
        cols_to_fill = ['ARIMA', 'Prophet', 'BiLSTM', 'Prophet_residuals', 'ARIMA_residuals']
        dataset[cols_to_fill] = dataset[cols_to_fill].fillna(method='ffill').fillna(method='bfill')
        
        # Apply scalers
        dataset['Close_scaled'] = scaler_close.fit_transform(dataset[['Close_original']])
        dataset['ARIMA_scaled'] = scaler_arima.fit_transform(dataset[['ARIMA']])
        dataset['Prophet_scaled'] = scaler_prophet.fit_transform(dataset[['Prophet']])
        dataset['BiLSTM_scaled'] = scaler_bilstm.fit_transform(dataset[['BiLSTM']])
        
        # Scale residuals (clipped)
        residuals_clipped_pro = dataset[['Prophet_residuals']].clip(-50, 50)
        dataset['Prophet_residuals_scaled'] = scaler_pro_resid.fit_transform(residuals_clipped_pro)
        
        residuals_clipped_ari = dataset[['ARIMA_residuals']].clip(-50, 50)
        dataset['ARIMA_residuals_scaled'] = scaler_ari_resid.fit_transform(residuals_clipped_ari)
        
        # Scale other technical indicators
        other_features = ['Returns', 'momentum', 'RSI', 'MACD', 'ema', 'ma7', 'ma21']
        scaler_others = MinMaxScaler()
        dataset[other_features] = scaler_others.fit_transform(dataset[other_features].fillna(0))
        
        # Store scalers
        self.scalers = {
            'close': scaler_close,
            'arima': scaler_arima,
            'prophet': scaler_prophet,
            'bilstm': scaler_bilstm,
            'pro_resid': scaler_pro_resid,
            'ari_resid': scaler_ari_resid,
            'others': scaler_others
        }
        
        return dataset
    
    def create_sequences(self, test_size=0.15):
        """Create sequences for prediction"""
        # Get clean data
        clean_df = self.weekly_df[self.input_features + ['Close_scaled']].dropna()
        
        X, y = [], []
        for i in range(self.window_size, len(clean_df)):
            X.append(clean_df[self.input_features].iloc[i-self.window_size:i].values)
            y.append(clean_df['Close_scaled'].iloc[i])
        
        X, y = np.array(X), np.array(y)
        
        # Split into train/test
        split_idx = int(len(X) * (1 - test_size))
        X_train, X_test = X[:split_idx], X[split_idx:]
        y_train, y_test = y[:split_idx], y[split_idx:]
        
        # Get corresponding dates
        dates = clean_df.index[self.window_size:]
        train_dates = dates[:split_idx]
        test_dates = dates[split_idx:]
        
        return (X_train, y_train, train_dates), (X_test, y_test, test_dates)
    
    def test_model(self):
        """Test the model and generate predictions"""
        if self.model is None:
            print("❌ Model not loaded!")
            return None
            
        print("🧪 Testing model...")
        
        # Create sequences
        (X_train, y_train, train_dates), (X_test, y_test, test_dates) = self.create_sequences()
        
        # Make predictions
        y_pred_train = self.model.predict(X_train).flatten()
        y_pred_test = self.model.predict(X_test).flatten()
        
        # Convert back to original scale
        y_train_orig = self.scalers['close'].inverse_transform(y_train.reshape(-1, 1)).flatten()
        y_pred_train_orig = self.scalers['close'].inverse_transform(y_pred_train.reshape(-1, 1)).flatten()
        
        y_test_orig = self.scalers['close'].inverse_transform(y_test.reshape(-1, 1)).flatten()
        y_pred_test_orig = self.scalers['close'].inverse_transform(y_pred_test.reshape(-1, 1)).flatten()
        
        # Calculate metrics
        train_metrics = self._calculate_metrics(y_train_orig, y_pred_train_orig, "Training")
        test_metrics = self._calculate_metrics(y_test_orig, y_pred_test_orig, "Testing")
        
        return {
            'train': (train_dates, y_train_orig, y_pred_train_orig),
            'test': (test_dates, y_test_orig, y_pred_test_orig),
            'train_metrics': train_metrics,
            'test_metrics': test_metrics
        }
    
    def _calculate_metrics(self, y_true, y_pred, dataset_name):
        """Calculate and print metrics"""
        mae = mean_absolute_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        mape = mean_absolute_percentage_error(y_true, y_pred) * 100
        r2 = r2_score(y_true, y_pred)
        
        print(f"\n📊 {dataset_name} Set Metrics:")
        print(f"   MAE  : ${mae:.2f}")
        print(f"   RMSE : ${rmse:.2f}")
        print(f"   MAPE : {mape:.2f}%")
        print(f"   R²   : {r2:.4f}")
        
        return {'MAE': mae, 'RMSE': rmse, 'MAPE': mape, 'R2': r2}
    
    def plot_results(self, results):
        """
        Create the main visualization that end users would see
        """
        if results is None:
            print("❌ No results to plot!")
            return
            
        train_dates, y_train_true, y_train_pred = results['train']
        test_dates, y_test_true, y_test_pred = results['test']
        
        # Create the main prediction plot
        fig, axes = plt.subplots(2, 2, figsize=(20, 12))
        fig.suptitle('Stock Price Prediction - Model Performance Dashboard', 
                    fontsize=20, fontweight='bold')
        
        # 1. Main Prediction Plot (what end users see)
        ax1 = axes[0, 0]
        ax1.plot(train_dates, y_train_true, label='Actual Price (Training)', color='#2E8B57', alpha=0.7, linewidth=1)
        ax1.plot(train_dates, y_train_pred, label='Predicted Price (Training)', color='#32CD32', alpha=0.8, linewidth=1)
        ax1.plot(test_dates, y_test_true, label='Actual Price (Testing)', color='#191970', linewidth=2)
        ax1.plot(test_dates, y_test_pred, label='Predicted Price (Testing)', color='#FF4500', linewidth=2, linestyle='--')
        
        # Add vertical line to separate train/test
        ax1.axvline(x=test_dates[0], color='red', linestyle=':', alpha=0.7, linewidth=2)
        ax1.text(test_dates[0], ax1.get_ylim()[1]*0.9, 'Test Period →', rotation=90, 
                fontsize=10, color='red', fontweight='bold')
        
        ax1.set_title('📈 Stock Price Prediction vs Reality', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Date', fontsize=12)
        ax1.set_ylabel('Stock Price ($)', fontsize=12)
        ax1.legend(loc='upper left', fontsize=10)
        ax1.grid(True, alpha=0.3)
        
        # 2. Test Period Focus (Zoomed view for end users)
        ax2 = axes[0, 1]
        ax2.plot(test_dates, y_test_true, label='Actual Price', color='#191970', linewidth=3, marker='o', markersize=4)
        ax2.plot(test_dates, y_test_pred, label='AI Prediction', color='#FF4500', linewidth=3, 
                linestyle='--', marker='s', markersize=4)
        
        ax2.set_title('🎯 Test Period: AI vs Reality', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Date', fontsize=12)
        ax2.set_ylabel('Stock Price ($)', fontsize=12)
        ax2.legend(fontsize=11)
        ax2.grid(True, alpha=0.3)
        
        # Add accuracy percentage
        accuracy = 100 - np.mean(np.abs((y_test_true - y_test_pred) / y_test_true)) * 100
        ax2.text(0.02, 0.98, f'Accuracy: {accuracy:.1f}%', transform=ax2.transAxes, 
                fontsize=12, fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen"),
                verticalalignment='top')
        
        # 3. Prediction Errors Over Time
        ax3 = axes[1, 0]
        errors = y_test_true - y_test_pred
        ax3.plot(test_dates, errors, color='purple', linewidth=2, marker='o', markersize=3)
        ax3.axhline(y=0, color='black', linestyle='--', alpha=0.5)
        ax3.fill_between(test_dates, errors, alpha=0.3, color='purple')
        
        ax3.set_title('📊 Prediction Errors Over Time', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Date', fontsize=12)
        ax3.set_ylabel('Error ($)', fontsize=12)
        ax3.grid(True, alpha=0.3)
        
        # Add error statistics
        mean_error = np.mean(errors)
        std_error = np.std(errors)
        ax3.text(0.02, 0.98, f'Mean Error: ${mean_error:.2f}\nStd Dev: ${std_error:.2f}', 
                transform=ax3.transAxes, fontsize=10, 
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightyellow"),
                verticalalignment='top')
        
        # 4. Performance Metrics Comparison
        ax4 = axes[1, 1]
        
        train_metrics = results['train_metrics']
        test_metrics = results['test_metrics']
        
        metrics = ['MAE', 'RMSE', 'MAPE', 'R2']
        train_values = [train_metrics[m] for m in metrics]
        test_values = [test_metrics[m] for m in metrics]
        
        x = np.arange(len(metrics))
        width = 0.35
        
        bars1 = ax4.bar(x - width/2, train_values, width, label='Training', color='skyblue', alpha=0.8)
        bars2 = ax4.bar(x + width/2, test_values, width, label='Testing', color='lightcoral', alpha=0.8)
        
        ax4.set_title('📋 Model Performance Metrics', fontsize=14, fontweight='bold')
        ax4.set_xlabel('Metrics', fontsize=12)
        ax4.set_ylabel('Values', fontsize=12)
        ax4.set_xticks(x)
        ax4.set_xticklabels(metrics)
        ax4.legend()
        ax4.grid(True, alpha=0.3, axis='y')
        
        # Add value labels on bars
        def autolabel(bars, values):
            for bar, value in zip(bars, values):
                height = bar.get_height()
                if abs(value) < 1:
                    label = f'{value:.3f}'
                else:
                    label = f'{value:.1f}'
                ax4.annotate(label, xy=(bar.get_x() + bar.get_width() / 2, height),
                           xytext=(0, 3), textcoords="offset points",
                           ha='center', va='bottom', fontsize=9)
        
        autolabel(bars1, train_values)
        autolabel(bars2, test_values)
        
        plt.tight_layout()
        plt.show()
        
        # Create a simplified "End User" view
        self._create_end_user_view(test_dates, y_test_true, y_test_pred, results['test_metrics'])
    
    def _create_end_user_view(self, test_dates, y_true, y_pred, metrics):
        """Create a clean, end-user focused visualization"""
        fig, ax = plt.subplots(1, 1, figsize=(15, 8))
        
        # Main plot
        ax.plot(test_dates, y_true, label='Actual Stock Price', color='#1f77b4', linewidth=3, marker='o', markersize=5)
        ax.plot(test_dates, y_pred, label='AI Prediction', color='#ff7f0e', linewidth=3, linestyle='--', marker='s', markersize=5)
        
        # Styling
        ax.set_title('Stock Price: AI Prediction vs Reality', fontsize=18, fontweight='bold', pad=20)
        ax.set_xlabel('Date', fontsize=14, fontweight='bold')
        ax.set_ylabel('Stock Price (USD)', fontsize=14, fontweight='bold')
        
        # Legend with better styling
        legend = ax.legend(loc='upper left', fontsize=12, frameon=True, fancybox=True, shadow=True)
        legend.get_frame().set_facecolor('white')
        legend.get_frame().set_alpha(0.9)
        
        # Grid
        ax.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)
        ax.set_facecolor('#f8f9fa')
        
        # Performance metrics box
        accuracy = 100 - metrics['MAPE']
        metrics_text = f"""📊 Model Performance:
        • Accuracy: {accuracy:.1f}%
        • Average Error: ${metrics['MAE']:.2f}
        • R² Score: {metrics['R2']:.3f}"""
        
        ax.text(0.02, 0.98, metrics_text, transform=ax.transAxes, fontsize=11,
                bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.8),
                verticalalignment='top', fontweight='bold')
        
        # Price range info
        current_price = y_true[-1]
        predicted_price = y_pred[-1]
        price_diff = predicted_price - current_price
        
        price_text = f"""💰 Latest Prediction:
        • Current: ${current_price:.2f}
        • Predicted: ${predicted_price:.2f}
        • Difference: ${price_diff:+.2f}"""
        
        ax.text(0.98, 0.02, price_text, transform=ax.transAxes, fontsize=11,
                bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgreen", alpha=0.8),
                verticalalignment='bottom', horizontalalignment='right', fontweight='bold')
        
        plt.tight_layout()
        plt.show()

    def predict_future(self, weeks_ahead=1):
        """Predict future prices for specified number of weeks ahead."""
        if not self.model:
            print("❌ Model not loaded!")
            return None
    
        clean_df = self.weekly_df[self.input_features + ['Close_scaled']].dropna()
        current_window = clean_df.iloc[-self.window_size:].copy()
        predictions = []
        
        for _ in range(weeks_ahead):
            # Prepare input
            X = current_window[self.input_features].values
            X = np.expand_dims(X, axis=0)
            
            # Predict
            pred_scaled = self.model.predict(X, verbose=0)[0][0]
            pred_price = self.scalers['close'].inverse_transform([[pred_scaled]])[0][0]
            predictions.append(pred_price)
            
            # Update window (simplified)
            new_row = current_window.iloc[-1:].copy()
            new_row['Close_scaled'] = pred_scaled
            current_window = pd.concat([current_window.iloc[1:], new_row])
    
        future_date = clean_df.index[-1] + pd.DateOffset(weeks=weeks_ahead)
        print(f"🔮 Predicted price for {future_date.strftime('%Y-%m-%d')}: ${predictions[-1]:.2f}")
        return future_date, predictions[-1]
    
    def predict_future_price(self, target_date_str):
        """
        Predict price for a specific future date
        """
        try:
            target_date = pd.to_datetime(target_date_str)

            # debug area
            print("Earliest date:", self.weekly_df.index.min())
            print("Latest date:", self.weekly_df.index.max())
            print("WEEKLY_TEST:", self.weekly_df.index)

            
            if target_date not in self.weekly_df.index:
                print(f"❌ Date {target_date_str} not found in dataset")
                return None
            
            idx = self.weekly_df.index.get_loc(target_date)
            
            if idx < self.window_size:
                print(f"❌ Not enough historical data for prediction")
                return None
            
            # Get feature window
            clean_df = self.weekly_df[self.input_features + ['Close_scaled']].dropna()
            target_idx = clean_df.index.get_loc(target_date)
            
            X_window = clean_df.iloc[target_idx - self.window_size:target_idx][self.input_features].values
            X_window = np.expand_dims(X_window, axis=0)
            
            # Predict
            pred_scaled = self.model.predict(X_window, verbose=0)
            pred_price = self.scalers['close'].inverse_transform(pred_scaled)[0][0]
            
            # Get actual price
            actual_price = self.weekly_df.loc[target_date, 'Close_original']
            
            print(f"📅 Date: {target_date_str}")
            print(f"✅ Actual Price: ${actual_price:.2f}")
            print(f"🤖 M Prediction: ${pred_price:.2f}")
            print(f"📊 Difference: ${pred_price - actual_price:+.2f}")
            
            return actual_price, pred_price
            
        except Exception as e:
            print(f"❌ Error in prediction: {e}")
            return None

def main():
    """
    Main function to run the independent model test
    """
    print("🚀 Starting Independent Model Test...")
    print("="*50)
    
    # Initialize tester (update paths as needed)
    tester = StockPredictionTester(
        model_path='stacked_bilstm.keras',  # Update this path
        data_path='datasets/aapl_us_d.csv'           # Update this path
        # data_path='datasets/clean_apple-hd.csv'           # Update this path
        
    )
    
    # Load model
    if not tester.load_model():
        print("❌ Failed to load model. Please check the model path.")
        return
    
    # Prepare data
    tester.prepare_data()
    
    # Test model
    results = tester.test_model()
    
    if results is None:
        print("❌ Testing failed!")
        return
    
    # Plot results (this is what end users would see)
    print("\n📈 Generating visualizations...")
    tester.plot_results(results)
    
    # Test specific date prediction
    print("\n🎯 Testing specific date prediction...")
    tester.predict_future_price('2024-09-06')  # Update date as needed
    
    print("\n✅ Testing completed successfully!")

if __name__ == "__main__":
    main()
```
